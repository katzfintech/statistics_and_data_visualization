```python
# Import libraries and load data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('stat-010_tips.csv')
df.head()
```

# Problem 1
Create the following 2 plots using Seaborn Python package and ‘tips’ dataset.<br>
Based on the given ‘tips’ dataset, who tips more? Males or females?


```python
# Problem 1-1
# Drow a barplot
bar = sns.catplot(data=df,x='sex',y='tip',height=3.5,kind='bar',order=['Male','Female'])
bar.set_ylabels('tip')
bar
```

```python
# Problem 1-2
# Drow a boxplot
box = sns.boxplot(data=df,x='day',y='tip',hue='sex',order=['Thur','Fri','Sat','Sun'],hue_order=['Male','Female'])
box
```

# Problem 2
Use Seaborn Python package to create the plots of this problem.<br>
Generate 2 sets of univariate data.
Univariate data set#1: 1000 normally distributed data with 𝜇 = 50, 𝜎 = 10<br>
Univariate data set#2: 1000 normally distributed data with 𝜇 = 100, 𝜎 = 50<br>
• Create an overlay histogram of set#1 and set#2 datasets. Next create a KDE (Kernel
Density Estimation) histogram by concatenating set#1 and set#2 datasets.<br>
• Create KDE Joint Plot<br>
• Create HEX Joint Plot<br>

```python
# Randomly generate datasets
rng = np.random.default_rng()
set1 = np.random.normal(loc=50,scale=10,size=1_000)
set2 = np.random.normal(loc=100,scale=50,size=1_000)
```

```python
# Concatenate datasets to create an overlay histgram + kde plot
fig, ax = plt.subplots()
ax = plt.hist([set1, set2],histtype = 'barstacked', density = True)
sns.kdeplot(np.concatenate((set1, set2)))
```

```python
# Create KDE Joint Plot
sns.jointplot(x=set1,y=set2,kind='kde',fill = True,space=0)
```

```python
# Create HEX Joint Plot
sns.jointplot(x=set1,y=set2,kind='hex',space=0)
```
