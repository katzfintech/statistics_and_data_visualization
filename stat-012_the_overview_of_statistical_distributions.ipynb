#1.Normal Distribution

```python
# Set rng to simulate
# rng = random number generator
rng = np.random.default_rng(seed=0)
```

```python
# Generate a sample normal distribution with 10,000 sample sizes
normal_sample = pd.DataFrame(rng.normal(loc=20, scale=10, size=100000), columns=['sample'])
normal_sample
```

```python
# Plot normal distribution
normal_sample.plot(kind='hist', bins=100, title='Sample Normal Distribution')
```

#2. Binomial Distribution
* E.g., coin toss (Bernoulli distribution)


```python
# Set rng to simulate
rng = np.random.default_rng(seed=1)
```

```python
# Generate 1,000,000 results of coin flips
bernoulli_sample = pd.DataFrame(rng.binomial(n=1, p=0.5, size=1000000), columns=['sample'])
bernoulli_sample
```

```python
# Plot coin flip results (Bernoulli distribution)
bernoulli_sample.plot(kind='hist', bins=100, title='Coin Flip Histogram')
```

```python
bernoulli_sample.value_counts().plot(kind='bar')
```

E.g., Categorical sample of Binomial distribution

```python
categorical_sample = pd.Series(rng.choice(['A', 'B', 'C'], p=[0.4, 0.5, 0.1], size=1000000))

categorical_sample.value_counts().sort_index()
```

```python
import matplotlib.pyplot as plt

categorical_sample.value_counts().sort_index().plot(kind='bar', title='Categorical Frequency Distribution')
plt.show()
```

# 3. Uniform Distribution

```python
# Set rng to simulate
rng = np.random.default_rng(seed=2)
```

```python
# Generate 100,000 samples from a uniform distribution between 1 and 3
uniform =pd.DataFrame(rng.uniform(low = 1,high= 3, size = 100000), columns = ['Samples'])
```

```python
# Plot a uniform distribution
uniform.plot(kind = 'hist', bins = 20, title ='Sample Uniform Distribution')
```

#4. Poisson Distribution

## Sample question
<br>
<font size="+1">
    <ul>
        <li>Assume customers are arriving at a grocery store according to some probability distribution over any given $10$ minute interval.</li>
        <br>
        <li>Further, assume the number of customers arriving in a $10$ minute interval is distributed according to a Poisson distribution with intesity rate of $\lambda = 8.5$.</li>
        <br>
        <li>Estimate the chance that the number of arrivals in a given $10$ minute interval is $15$ or higher.</li>
        <br>
    </ul>
</font>

$\square$

```python
# Set rng to simulate
rng = np.random.default_rng(seed=3)
```

```python
# Generate 1,000,000 samples from a poisson distribution whose lamnda = 8.5
poisson_sample = rng.poisson(lam=8.5, size =1000000)
```

```python
# Calculate the probability
print(f'The probability of the sampling is: {round(np.mean(poisson_sample >= 15),3)}')
```

```python
# Plot a poisson distribution
poisson_sample = pd.DataFrame(rng.poisson(lam=8.5, size =1000000), columns=['sample'])

poisson_sample.plot(kind='hist', bins=100, title='Sample Poisson Distribution')
```

# 5. Geometric Distribution

```python
# Set rng to simulate
rng = np.random.default_rng(seed=4)
```

```python
# Simulate  100,000  samples from a geometric distribution with parameter 𝑝=0.3
geometric_sample =pd.DataFrame(rng.geometric(0.3, size=100000), columns = ['Samples'])
```

```python
# Plot a geometric distribution
geometric_sample.plot(kind = 'hist', bins = 30, title ='Geometric Sampling Distribution')
```
