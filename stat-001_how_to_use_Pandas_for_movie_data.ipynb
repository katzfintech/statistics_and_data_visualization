```python
# Question 1
# Read the ‘top_movies_by_title.csv’ file into a Pandas’ data frame and name it ‘table1’
import pandas as pd
table1 = pd.read_csv('stat-001_top_movies_by_title.csv')
table1.head()
```

```python
# Question 2
# Select ‘Title’ and ‘Year’ of the ‘table1’, into another table and name is ‘table2’
table2 = table1[['Title','Year']]
table2
```

```python
# Question 3
# Select only the movies from ‘table1’ where the gross revenues generated by the movie exceeds 100 million.
over_100m = table1[table1['Gross']>= 100000000]
over_100m
```

```python
# Question 4
# Select only the movies from ‘table1’ where the gross revenues generated by the movie is
# between 100 and 200 million. Sort the output data frame by 'Gross' revenues in descending
# order. Display only the top 10 entries.

over_100m_under_200m = table1[(table1['Gross']>= 100000000) & (table1['Gross']<= 200000000)].sort_values(by=['Gross'], ascending=False)
over_100m_under_200m.head(10)
```

```python
# Question 5
# Select movies where the word ‘the’ appears in the title of the movie. Show the first 10 movies

movie_incl_the = table1[table1['Title'].str.contains('the')]
movie_incl_the.head(10)
```

```python
# Question 6-1
# Sort the movies in ‘table1’ in descending order (highest value first) by their ‘Gross’ revenues.
# Select top 3 movies made the most amount of money based on ‘Gross’ column.

gross = table1.sort_values('Gross', ascending=False)
gross.head(3)
```

```python
# Question 6-2
# Select top 3 movies made the most amount of money based on ‘Gross (Adjusted)’ column.
gross_adj = table1.sort_values('Gross (Adjusted)', ascending=False)
gross_adj.head(3)
```

```python
# Question 7
# Count the number of movies from each studio. How many movies made by ‘Columbia’ and ‘Fox’ studios?

studio = table1['Studio'].value_counts()
studio

#Answer: Columbia = 10, Fox = 26
```

```python
# Question 8
# Which is the highest grossing movie from ‘Columbia’ studio?

columbia = table1[table1['Studio'] == 'Columbia'].sort_values('Gross', ascending=False)
columbia.head(1)
```

```python
# Question 9
# Compute the ratio between ‘Gross (Adjusted)’ and ‘Gross’ for all movies. Next sort the table by
# the ratio in descending order. Display the top 10 ratios
table1['Ratio'] = table1['Gross (Adjusted)'] / table1['Gross']
table1.sort_values('Ratio', ascending=False).head(10)
```

```python
# Question 10
# Group all the movies by the year in which they were released. In which year has the maximum
# number of movies released?

year = table1['Year'].value_counts()
year

#Answer:2002,2004,2009,2015
```
