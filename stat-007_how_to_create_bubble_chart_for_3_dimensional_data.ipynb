# 1.2<br>
Compute the “average life expectancy” of males and females for all the countries. We will use this
variable in the next 2 problems (1.3 + 1.4).

```python
# Compute the average life expectancy
data['Avg Life Exp'] = (data['Life Exp Female'] + data['Life Exp Male'])/2
data.head()
```

# 1.3<br>
Create 2 scatter bubble charts between GDP and “average life expectancy” for all the countries in
the dataset.<br>
• In the first chart the size of the bubble should be proportional to the population of the country.<br>
• In the second chart the size of the bubble should be proportional to the surface area of the
country.

```python
# Import libraries
import matplotlib.pyplot as plt

# Create the first chart
x = data['GDP (M US$)']
y = data['Avg Life Exp']
pop = data['Population 1,000  (2017)']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cm = plt.cm.get_cmap('prism')
scat = ax.scatter(x,y,c=pop,s=pop/1000,alpha=0.5,cmap=cm)
plt.title('Relationship between GDP and Average Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Create the second chart

x = data['GDP (M US$)']
y = data['Avg Life Exp']
size = data['Surface area (km2)']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cm = plt.cm.get_cmap('prism')
scat = ax.scatter(x,y,c=size,s=size/10000,alpha=0.5,cmap=cm)
plt.title('Relationship between GDP and Average Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```

# 1.4<br>
The plots created in the previous problem (1.3) have bubbles squeezed on the left side of the
graph. Create 2 more scatter bubble charts between ‘log (natural log) value of GDP’ and “average
life expectancy” for all the countries in the dataset. This will allow the bubbles to spread out all over the graph.<br>
• In the first chart the size of the bubble should be proportional to the population of the country.<br>
• In the second chart the size of the bubble should be proportional to the surface area of the
country.<br>

```python
# Import libraries
import numpy as np

# Create the first chart
x = np.log(data['GDP (M US$)'])
y = data['Avg Life Exp']
pop = data['Population 1,000  (2017)']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cm = plt.cm.get_cmap('prism')
scat = ax.scatter(x,y,c=pop,s=pop/1000,alpha=0.5,cmap=cm)
plt.title('Relationship between GDP and Average Life Expectancy')
plt.xlabel('Log GDP')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Create the second chart
x = np.log(data['GDP (M US$)'])
y = data['Avg Life Exp']
size = data['Surface area (km2)']
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cm = plt.cm.get_cmap('prism')
scat = ax.scatter(x,y,c=size,s=size/10000,alpha=0.5,cmap=cm)
plt.title('Relationship between GDP and Average Life Expectancy')
plt.xlabel('Log GDP')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```
