Plot the following 3 functions.

2. ùë¶ = 40(1.7)ùë• ùê∏ùë•ùëùùëúùëõùëíùëõùë°ùëñùëéùëô ùêπùë¢ùëõùëêùë°ùëñùëúùëõ
3. ùë¶ = 15ùë•0.37 ùëÉùëúùë§ùëíùëü ùêπùë¢ùëõùëêùë°ùëñùëúùëõ
Plot these 3 functions in the following format:<br>
a. Rectilinear scale<br>
b. Semi log scale: y-scale is log (natural log)<br>
c. Log/Log scale: both x and y scales are log (natural log)
<br>
<br>
You will notice that the exponential functions will generate a straight line in semi-log scale.<br>
And power functions will generate a straight line in log-log scale.<br>
<br>
Make sure that your ‚Äòx‚Äô variable should not include zero because the log(0) is minus infinity.

```python
# 1. ùë¶ = 25 ‚àó ùëí0.5ùë•

# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Define x and y
x = np.linspace(0, 10,100)
y = 25 * np.exp(0.5 * x)

# a. Rectilinear scale
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rectilinear scale')
plt.legend()
plt.show()
```

```python
# b Semi log scale
plt.semilogy(x, y)
plt.xlabel('x')
plt.ylabel('log(y)')
plt.title('Semi log scale')
plt.legend()
plt.show()
```

```python
# c Log/Log scale
plt.loglog(x, y)
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title('Log/Log scale')
plt.legend()
plt.show()
```

```python
# 2. ùë¶ = 40(1.7)ùë•

# Define x and y
x = np.linspace(0, 10,100)
y = 40 * 1.7**x

# a. Rectilinear scale
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rectilinear scale')
plt.legend()
plt.show()
```

```python
# b Semi log scale
plt.semilogy(x, y)
plt.xlabel('x')
plt.ylabel('log(y)')
plt.title('Semi log scale')
plt.legend()
plt.show()
```

```python
# c Log/Log scale
plt.loglog(x, y)
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title('Log/Log scale')
plt.legend()
plt.show()
```

```python
# 3. ùë¶ = 15ùë•0.37

# Define x and y
x = np.linspace(0, 10,100)
y = 15 * x ** 0.37

# a. Rectilinear scale
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Rectilinear scale')
plt.legend()
plt.show()
```

```python
# b Semi log scale
plt.semilogy(x, y)
plt.xlabel('x')
plt.ylabel('log(y)')
plt.title('Semi log scale')
plt.legend()
plt.show()
```

```python
# c Log/Log scale
plt.loglog(x, y)
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title('Log/Log scale')
plt.legend()
plt.show()
```

# Problem #2
Use Matplotlib‚Äôs Object-Oriented API to create a plot with the following characteristics.<br>
* First create ‚Äòfigure‚Äô and ‚Äòaxes‚Äô objects, and then populate these objects with data and then create a plot.<br>
* Data: All trigonometric functions: sine, cosine, tangent, cosecant, secant, cotangent. The domain of these functions should be from 0 (zero) to 2œÄ.
* Create 3 subplots to show different functions.
 - Sine/Cosecant
 - Cosine/Secant
 - Tangent/Cotangent
* Set the figure size of 20-centimeter wide and 30-centimeters high.
* Add labels on the x-axis and y-axis on all sub-plots.
* Add annotation in the plot which describe the type of plot.
* Add a Legend in all plots.
* Save plot as an image.
* Specify Axis range (both x and y range).
* Specify tick frequency as displayed on the plot.

```python
import math

x = np.linspace(0, 2 * np.pi, 1000)
x_90 = x[(x != math.pi/2) & (x != math.pi*3/2)]
x_pi = x[x % math.pi != 0]
y_sine = np.sin(x)
y_cosecant = 1 / np.sin(x_pi)
y_cosine = np.cos(x)
y_secant = 1 / np.cos(x_90)
y_tangent = np.tan(x_90)
y_cotangent = 1 /np.tan(x_pi)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(20, 30))

axes[0].plot(x, y_sine, label = 'sine')
axes[0].plot(x_pi, y_cosecant, label = 'cosecant')
axes[0].set_title('Sine and Cosecant')
axes[0].legend()

axes[1].plot(x, y_cosine, label = 'cosine')
axes[1].plot(x_90, y_secant, label = 'secant')
axes[1].set_title('Cosine and Secant')
axes[1].legend()

axes[2].plot(x_90, y_tangent, label = 'tangent')
axes[2].plot(x_pi, y_cotangent, label = 'cotangent')
axes[2].set_title('Tangent and Cotangent')
axes[2].legend()

for ax in axes.flatten():
    for j in ['bottom', 'top', 'right', 'left']:
        ax.spines[j].set_position(('data', 0))
    ax.set_xlim(0, 2*math.pi)
    ax.set_xticks(np.linspace(-2 * np.pi, 2 * np.pi, 9, endpoint = True))
    ax.set_xticklabels(['-2œÄ', '-3œÄ/2', '-œÄ', '-œÄ/2', '0', 'œÄ/2', 'œÄ', '3œÄ/2', '2œÄ'])
    ax.set_ylim(-3, 3)
    ax.set_yticks([-3, -2, -1, 0, 1, 2, 3])
    ax.set_yticklabels([-3, -2, -1, 0, '1', '2', '3'])

    ax.legend()

fig.savefig('Graphs.png', dpi=300)
```

# Problem #3
The Virginia Cooperative Extension reports that the mean weight of yearling Angus steers is 1,152 pounds.<br>
Suppose that weights of all such animals can be described by a Normal model with a standard deviation of 84 pounds.<br>
Plot the normal distribution curve (¬µ = 1,152 and œÉ = 84). Fill the distribution curve with your choice of color:<br>
1. Where weight is over 1,250 pounds
2. Where weight is under 1,200 pounds
3. Where weight is between 1,000 and 1,100 pounds

```python
# Import liblaries
import scipy.stats as stats
from scipy.stats import norm

# 1.Where weight is over 1,250 pounds
x = np.arange(800,1500,0.01)
fig, ax = plt.subplots()

ax.plot(x,norm.pdf(x,1152,84))
ax.set_xlabel('z')
ax.set_ylabel('pd')

ax.fill_between(x=np.arange(1250,1500,0.01),
                y1=stats.norm.pdf(np.arange(1250,1500,0.01), loc = 1152, scale = 84),
                facecolor='orange',
                alpha = 1.0)

plt.show()
```

```python
# 2. Where weight is under 1,200 pounds
x = np.arange(800,1500,0.01)
fig, ax = plt.subplots()

ax.plot(x,norm.pdf(x,1152,84))
ax.set_xlabel('z')
ax.set_ylabel('pd')

ax.fill_between(x=np.arange(800,1200,0.01),
                y1=stats.norm.pdf(np.arange(800,1200,0.01), loc = 1152, scale = 84),
                facecolor='orange',
                alpha = 1.0)

plt.show()
```

```python
# 3. Where weight is between 1,000 and 1,100 pounds
x = np.arange(800,1500,0.01)
fig, ax = plt.subplots()

ax.plot(x,norm.pdf(x,1152,84))
ax.set_xlabel('z')
ax.set_ylabel('pd')

ax.fill_between(x=np.arange(1000,1100,0.01),
                y1=stats.norm.pdf(np.arange(1000,1100,0.01), loc = 1152, scale = 84),
                facecolor='orange',
                alpha = 1.0)

plt.show()
```
