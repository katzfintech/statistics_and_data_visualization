```python
# Problem 1
# The file Energy contains the per capita energy consumption,
# in kilowatt-hours, for each of the 50 states and the District of
# Columbia during a recent year.

# Problem 1-(a)
# Compute the mean, variance, and standard deviation for the population.

import pandas as pd
ene = pd.read_csv('stat-003-1_Energy.csv')

# calucurate each value
hours_mean = ene['Kilowatt Hours'].mean()
hours_var = ene['Kilowatt Hours'].var(ddof = 0)
hours_std = ene['Kilowatt Hours'].std(ddof = 0)

print(hours_mean)
print(hours_var)
print(hours_std)
```

```python
# Problem 1-(b)
# What proportion of these states has per capita energy consumption within
# +-1 standard deviation of the mean, within +-2 standard deviations of the mean,
# and +-3 standard deviations of the mean?

print(f"The range of +-1 standard deviation of the mean is from {round(hours_mean-hours_std,2)} to {round(hours_mean+hours_std,2)}.")
print(f"The range of +-2 standard deviation of the mean is from {round(hours_mean-2*hours_std,2)} to {round(hours_mean+2*hours_std,2)}.")
print(f"The range of +-3 standard deviation of the mean is from {round(hours_mean-3*hours_std,2)} to {round(hours_mean+3*hours_std,2)}.")

# The proportion +-1 standard deviation of the mean
one_sigma_range = ene['Kilowatt Hours'][(ene['Kilowatt Hours']>= (hours_mean-hours_std)) & (ene['Kilowatt Hours']<= hours_mean+hours_std)].count()
proportion_one_sigma = one_sigma_range / ene['Kilowatt Hours'].count()
print(f"The proportion of +-1 standard deviation of the mean is {round(proportion_one_sigma,3)}.")

# The proportion +-2 standard deviation of the mean
one_sigma_range = ene['Kilowatt Hours'][(ene['Kilowatt Hours']>= (hours_mean-2*hours_std)) & (ene['Kilowatt Hours']<= hours_mean+2*hours_std)].count()
proportion_one_sigma = one_sigma_range / ene['Kilowatt Hours'].count()
print(f"The proportion of +-2 standard deviation of the mean is {round(proportion_one_sigma,3)}.")

# The proportion +-3 standard deviation of the mean
one_sigma_range = ene['Kilowatt Hours'][(ene['Kilowatt Hours']>= (hours_mean-3*hours_std)) & (ene['Kilowatt Hours']<= hours_mean+3*hours_std)].count()
proportion_one_sigma = one_sigma_range / ene['Kilowatt Hours'].count()
print(f"The proportion of +-3 standard deviation of the mean is {round(proportion_one_sigma,3)}.")
```

```python
# Problem 1-(c)
# Compare your findings with what would be expected based on
# the empirical rule. Are you surprized at the results in (b)

# My answer:
# Compared to the empirical rule, the differences of the result are as follows.
#  +-1 standard deviation: 68%(empirical) > 64.7%(actual) Difference: 3.3%
#  +-2 standard deviation: 95%(empirical) < 98.0%(actual) Difference: 3.0%
#  +-3 standard deviation: 99.7%(empirical) < 100.0%(actual) Difference: 0.3%

# Personally, it is not suprising because the errors are only around 3%, which is close to the Empirical rule.
```

```python
# Problem #9
# Toby's Trucking Company determined that the distance traveled per truck per year
# is normally distributed, with a mean of 50 thousand miles and a standard deviation of
# 12 thousand miles.

# Problem 9-(a)
# What proportion of tricks can be expected to trabel between 34 and 50 thousand miles in a year?

# Define variables
mu = 50
sigma = 12

# Z-score standardization
right_end_a = (34 - mu) / sigma
left_end_a = (50 - mu) / sigma

p9a = norm.cdf(left_end_a) - norm.cdf(right_end_a)

print(p9a)
```

```python
# Problem 9-(b)
# What percentage of trucks can be expected to travel either less than 30 or more than 60 thousand miles in a year?

right_end_b = (30 - mu) / sigma
left_end_b = (60 - mu) / sigma

p9b = norm.cdf(right_end_b) + (1-norm.cdf(left_end_b))
print(p9b)
```

```python
# Problem 9-(c)
# How many miles will be traveled by at least 80% of the trucks?
z_score = norm.isf(q = 0.20)
p9c = z_score * sigma + mu
print(p9c)
```

```python
# Problem 9-(d)
# What are your answers to (a) through (c) if the standard deviation is 10 thousand miles?

# Let's change the sigma
# Define variables
mu2 = 50
sigma2 = 10

#(a) Z-score standardization
right_end_da = (34 - mu2) / sigma2
left_end_da = (50 - mu2) / sigma2
p9da = norm.cdf(left_end_da) - norm.cdf(right_end_da)
print(f"The answer of (a) is changed to {p9da}")

# (b)Z-score standardization
right_end_db = (30 - mu2) / sigma2
left_end_db = (60 - mu2) / sigma2

p9db = norm.cdf(right_end_db) + (1-norm.cdf(left_end_db))
print(f"The answer of (b) is changed to {p9db}")

#(c)
z_score = norm.isf(q = 0.20)
p9dc = z_score * sigma2 + mu2
print(f"The answer of (c) is changed to {p9dc}")
```

```python
# Problem #10
# A set of final examination grades in an introductory statistics course is normally distributed,
# with a mean of 73 and a standard deviation of 8.

# Problem 10-(a)
# What is the probability that a student scored below 91 on this exam?

# Define variables
p10_mu = 73
p10_sigma = 8

#Z-score Standardization
value = (91 - p10_mu)/p10_sigma
p10a = norm.cdf(value)
print(p10a)
```

```python
# Problem 10-(b)
# What is the probability that a student scored between 65 and 89?

#Z-score Standardization
right_end_p10 = (65 - p10_mu)/p10_sigma
left_end_p10 = (89 - p10_mu)/p10_sigma
p10b = norm.cdf(left_end_p10) - norm.cdf(right_end_p10)
print(p10b)
```

```python
# Problem 10-(c)
# The probability is 5% that a student taking the test scores higher than what grade?
z_score_p10 = norm.isf(q = 0.05)
p10c = z_score_p10 * p10_sigma + p10_mu
print(p10c)
```

```python
# Problem #11
# The file SUV contains the overall miles per gallon of 2013 small SUVs

#Decide whether the data appear to be approximately normally distributed by...
# a. comparing data characteristics to theoretical properties.
# b. constructing a normal probability plot.

# Import libraries
import numpy as np
import pylab
import scipy.stats as stats
import matplotlib.pyplot as plt

# Read the csv file
suv = pd.read_csv('stat-003-4_SUV.csv')

# Check the following things.
suv_mean = suv['MPG'].mean()
suv_median = suv['MPG'].median()
suv_max = suv['MPG'].max()
suv_min = suv['MPG'].min()
suv_std = suv['MPG'].std()
suv_skew = suv['MPG'].skew()

print(f"Mean = {suv_mean}\nMedian = {suv_median}\nMax = {suv_max}\nMin = {suv_min}\nStd = {suv_std}\nSkewness = {suv_skew}")

# Sort the data
suv = suv.sort_values(by="MPG", ascending = True)

# Create the normal probability plot
stats.probplot(suv['MPG'], dist="norm", plot=pylab)
pylab.show

# Final answer:
# Based on the analysis, the data are NOT approximately normally distributed.
# In fact, the data are positively skewed according to the skewness and the probability plot.
```
